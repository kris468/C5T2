## Data Pre-Processing

    Discretize Limit_Bal and Age features for easier interpretation later.

```python
creditLimits = ["Small","Medium","Large","Super"]
credit["LIMIT_BAL_CAT"] = pd.cut(credit["LIMIT_BAL"],[9999,50000,250000,500000,1000001],labels = creditLimits)
creditAge = ["YoungAdults","MiddleAge","Elders"]
credit["AGE_CAT"] = pd.cut(credit["AGE"],[20,35,55,80],labels = creditAge)
```

```python
credit.groupby('LIMIT_BAL_CAT')['LIMIT_BAL_CAT'].count()
```

```python
credit.groupby('AGE_CAT')['AGE_CAT'].count()
```

Majority of customers are under age of 55 (almost 97.2 percent).

```python
credit.groupby('EDUCATION')['EDUCATION'].count()
```

Customers with some form of education (high school and beyond) comprise more than 98 percent of loans.

```python
#Distribution of Limit balance grouped by default
fig = sns.FacetGrid(credit, hue='default', aspect=4)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit.LIMIT_BAL.max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Limit Balance Grouped by Default')
fig.set(label=('No Default', 'Default'))
fig.add_legend()
```

Age doesnt seem to have a significant bearing on the tendency to default or not default a loan payment.
